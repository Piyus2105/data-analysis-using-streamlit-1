```python
import streamlit as st
import pandas as pd
import re
import plotly.express as px


st.set_page_config(layout='wide', page_title='Dashboard')

@st.cache_data
def load_data():
  url = 'https://raw.githubusercontent.com/skathirmani/datasets/refs/heads/main/narendramodi_tweets.csv'
  data = pd.read_csv(url)
  data['created_at']=pd.to_datetime(data['created_at'])
  data['Year'] = data['created_at'].dt.year
  return data
data=load_data()

st.sidebar.header('Narendra Modi - Twitter Analysis by Piyush kr. Agrawal')

year_filter= st.sidebar.slider(
    label='Select a year',
    min_value=2016,
    max_value=2017,
    step=2
    )


source_filter = st.sidebar.selectbox(
  label='source',
  options= data['source'].unique())

filtered_data=data[(data['Year']==(year_filter)) & (data['source']==source_filter)]

avg_likes=filtered_data['favorite_count'].mean()
avg_likes
```
